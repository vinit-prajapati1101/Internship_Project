```python
#Importing the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy  import linkage,dendrogram
from sklearn.metrics import silhouette_score
```

```python
#Load/Read the datset
df=pd.read_csv(r"C:\Users\VINIT\Wholesale customers data.csv")
df
```

```python
#Assign the df into variable x
x=df[['Grocery','Detergents_Paper']]
```

```python
x
```

```python
#Linkage the x and assign the method
l = linkage(x,method='ward')
l
```

```python
#Plotting the Dendrogram graph
plt.figure(figsize=(22,12))
dendrogram(l)
plt.show()
```

```python
#Create the model
model = AgglomerativeClustering(n_clusters = 3)
```

```python
#Train and test the data
p= model.fit_predict(x)
p
```

```python
#silhouette Score of data
sc=silhouette_score(x,p)
sc
```
