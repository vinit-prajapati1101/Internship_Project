```python
#Importing the Libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
#Load/Read the datasets
df = pd.read_csv(r"C:\Users\VINIT\sales-selected-columns.csv")
df
```

```python
#Define the LabelEncoder
le = LabelEncoder()
df.columns
```

```python
#Transform gender(categorical) to gender(numeric)
df['gender']= le.fit_transform(df['gender'])
df
```

```python
#Independent Variable(x)
x = df[['quantity']]
x
```

```python
#Dependent Variable(y)
y = df['tax']
y
```

```python
#Splitting the data
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Create the Model
model = LinearRegression()
```

```python
#Train the Model
model.fit(x_train,y_train)
```

```python
#Predict the result
y_pred = model.predict(x_test)
y_pred
```

```python
#Score of training model
model.score(x_train,y_train)
```

```python
#Score of testing model
model.score(x_test,y_test)
```

```python
#Intercept
i=model.intercept_
i
```

```python
#coefficient
c=model.coef_
c
```
